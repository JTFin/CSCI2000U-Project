```python
import csv
import re
from functools import reduce
import numpy as np

def get_data_csv():
    collection = []
    with open('imdb.csv', 'r') as f:
        for line in csv.DictReader(f):
            collection.append(line)
        return collection
        
# the data    
data = get_data_csv()
```

```python
import json
import pandas as pd

df = pd.DataFrame(data)
df = df.loc[df['Rate']!= 'No Rate']
df["Rate"] = pd.to_numeric(df["Rate"], downcast="float")
df_pivot = pd.pivot_table(data=df, values='Rate', index=['Date'], columns=['Type'], aggfunc='mean')
df_pivot.plot(kind='line', figsize = (10,5))
```

```python
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Violence'], columns=df['Type'], aggfunc='mean')
df_pivot.plot(kind='bar')
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Nudity'], columns=df['Type'], aggfunc='mean')
df_pivot.plot(kind='bar')
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Profanity'], columns=df['Type'], aggfunc='mean')
df_pivot.plot(kind='bar')
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Alcohol'], columns=df['Type'], aggfunc='mean')
df_pivot.plot(kind='bar')
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Frightening'], columns=df['Type'], aggfunc='mean')
df_pivot.plot(kind='bar')
```

```python
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Date'], columns=df['Frightening'], aggfunc='count')
df_pivot.plot(kind='line', figsize = (10,5))
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Date'], columns=df['Violence'], aggfunc='count')
df_pivot.plot(kind='line', figsize = (10,5))
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Date'], columns=df['Nudity'], aggfunc='count')
df_pivot.plot(kind='line', figsize = (10,5))
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Date'], columns=df['Profanity'], aggfunc='count')
df_pivot.plot(kind='line', figsize = (10,5))
df_pivot = pd.pivot_table(data=df, values='Rate', index=df['Date'], columns=df['Alcohol'], aggfunc='count')
df_pivot.plot(kind='line', figsize = (10,5))
```

```python
df_pivot = pd.pivot_table(data=df, values='Rate', index=['Duration','Episodes'], columns=['Type'], aggfunc='mean')
df_pivot.plot(kind='line', figsize = (15,8))
```
